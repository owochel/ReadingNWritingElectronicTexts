```python
import tracery
from tracery.modifiers import base_english

ingrident={
    "origin":
    """#ingredient#""",
    "ingredient":["#meat#","#veg#","#meat# #base#","#veg# #base#","#adj# #meat# #base#","#adj# #veg# #base#","#meat# #organ#","#veg# #organ#"],
    "meat":["#reds#","#flys#","#swims#"],
    "veg":["#starch#","#roots#","#greens#","#flower#","#nuts#","#fruits#"],
    "other":["#adj# #sea#","#adj# #sea# #base#","#sea#","#sea# #base#","#sea# #organ#","#oth#","#meat# #oth#"],
    "base":base,
    "adj":adj,
    "reds":reds,
    "flys":flys,
    "roots":roots,
    "swims":swims,
    "starch":starch,
    "greens":greens,
    "flower":flower,
    "nuts":nuts,
    "fruits":fruits,
    "organ":organ,
    "sea":seasoning, 
    "oth":other
}
grammar = tracery.Grammar(ingrident)
grammar.add_modifiers(base_english)
```

```python
import json
import os
from random import randrange

foods=[]
for i in range(1000):
    baseNum=int(randrange(1,15))
    otherNum=int(randrange(10))
    ingredients=[]
    for i in range(baseNum):
        ingredients.append(grammar.flatten("#ingredient#"))
    for i in range(otherNum):
        ingredients.append(grammar.flatten("#other#"))
    ingredients=[*set(ingredients)]#remove duplicates
    
    servingSize=random.randrange(1,100)
    servings=round(random.random()*10,1)
    nutrition=[]
    fat=random.randrange(0,servingSize)
    carbohydrate=random.randrange(servingSize - fat)
    protein=servingSize-fat-carbohydrate
    cholesterol=random.randrange(0,300)
    sodium=random.randrange(0,2300)


    fatPercent=round(float(fat/44)*100,2)
    carbohydratePercent=round(float(carbohydrate/45*100),2)
    proteinPercent=round(float(protein/56)*100,2)
    sodiumPercent=round(float(sodium/2300*100),2)
    cholesterolPercent=round(float(cholesterol/300*100),2)
    calories= 9*fat+4*(carbohydrate+protein)
    if fat!=0:
        satFat=random.randrange(0,fat)
        if carbohydrate!=0:
            fib=random.randrange(0,carbohydrate)
            ts=carbohydrate-fib-random.randrange(0,carbohydrate-fib)
            foods.append(json.dumps({
            "calories":calories,
            "servingSize":servingSize,
            "servings":servings,
            "fat":fat,
            "saturatedFat":satFat,
            "transFat":fat-satFat-random.randrange(0,fat-satFat),
            "fatPercent":fatPercent,
            "carbohydrate":carbohydrate,
            "carbohydratePercent":carbohydratePercent,
            "dietaryFiber":fib,
            "totalSugar":ts,
            "addedSugar":random.randrange(0,ts),
            "protein":protein,
            "proteinPercent":proteinPercent,
            "cholesterol":cholesterol,
            "cholesterolPercent":cholesterolPercent,
            "sodium":sodium,
            "sodiumPercent":sodiumPercent,
            "ingredients":ingredients,
            }))
        else:
            foods.append(json.dumps({
            "calories":calories,
            "servingSize":servingSize,
            "servings":servings,
            "fat":fat,
            "saturatedFat":satFat,
            "transFat":fat-satFat-random.randrange(0,fat-satFat),
            "fatPercent":fatPercent,
            "carbohydrate":carbohydrate,
            "carbohydratePercent":carbohydratePercent,
            "protein":protein,
            "proteinPercent":proteinPercent,
            "cholesterol":cholesterol,
            "cholesterolPercent":cholesterolPercent,
            "sodium":sodium,
            "sodiumPercent":sodiumPercent,
            "ingredients":ingredients,
            }))
    else:
        foods.append(json.dumps({
        "calories":calories,
        "servingSize":servingSize,
        "servings":servings,
        "fat":fat,
        "fatPercent":fatPercent,
        "carbohydrate":carbohydrate,
        "carbohydratePercent":carbohydratePercent,
        "protein":protein,
        "proteinPercent":proteinPercent,
        "cholesterol":cholesterol,
        "cholesterolPercent":cholesterolPercent,
        "sodium":sodium,
        "sodiumPercent":sodiumPercent,
        "ingredients":ingredients,
    }))
# os.remove('foods.json')

with open('foods.json', 'w') as f:
    f.write("{")
    f.write(" \"foods\":")
    f.write("[")
    for i in range(len(foods)):
        f.write(foods[i])
        if i<len(foods)-1:
            f.write(",")
    f.write("]")
    f.write("}")

    
    
```
