# Nutrition Facts Generator
I struggled with relationships with food for a long time in undergraduate and even still a bit at this moment. There was a period when I had to actively look at what and how others are eating to elect my food and pace myself because what seemed so natural to others regarding eating was not simply not registering in my mind. I was able to calculate the calories of a plate-full of food in seconds at glance. I stared at the nutrition facts of packages for hours in guilts and remorse. And the worst part is that I am a really foodie person by nature - I love to cook and learn new recipes, try new flavors and experience novel food, and greatly enjoy eating in general. Therefore, my troubled food relationship took a major part of joy out of my life.

Making this generator, I was able to do the design and calculation without additional research and reference because the layout and "suggested" values were sealed into my brain (I still searched to varified and no correction was needed). It is because of how the piece originated from such intimate personal experience, instead of treating the final result as a piece of work, I hope to emphasize the creation process for this piece. As a result, I used print() instead of a saved value or tracery for this creation. The generated nutrition fact table is directly printed out on the output window and supposedly left no marks after the next run like how I hope I could view a nutrition fact table when I was troubled by my relationship with food.Going off the ingredients list generator, I'm adding on another aspect to this food project by introducing the ingredients list of this food generator.

The intredients are listed based on my knowledge and seperated by types, and rearranged using language logic and my interpretation. The products of this project is limited by 
a) the pool selection of origin words - they ara based on my perception and understanding of edible ingredients;
b) the types division is based on my research and perception of commonly seen ingredients types, so the taxtonomy is only reflective to my observation;
c) the rearranging of the words is based on my understanding on normal food combination.
By no means, the construct of this project is intended to be complete and reflective for the real world as they are only a reflection of parts of my perception, understanding of foods in my world.

```python
import random
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'

print(color.BOLD + 'Hello World !' + color.END)
#with help from https://stackoverflow.com/questions/8924173/how-do-i-print-bold-text-in-python
```

```python
base=[
    'meat',
    'flour',
    'syrup',
    'water',
    'starch',
    'oil',
    'fat',
    'milk',
    'powder',
    'concentrate',
    'extract',
    'protein isolate',
    'puree',
    'cheese',
    'blood',
    'juice',
    'egg',    
    'shell',
    'flavor',
    'acid'
]

reds=['beef','cow','lamb','rabbit','mutton','pig','pork','venison','deer','goat','bacon','sausage','ham','salami','prosciutto','camel','llama']
flys=['chicken','goose','quail','duck','turkey','pigeon','ostrich','bat','ortolan','hummingbird','blue jay']
swims=['fish','shrimp','crab','craw fish','shellfish','whale','shark','salmon','catfish','loach','crocodile']

starch=['pea','hemp','chickpea','bean','lentil','wheat','rice','corn','soy','oat','malt']
roots=['potato','sweet potato','yam','carrot','beet','bamboo shoot','tarnip','onion','garlic','yuca','radish','ginger']
greens=['leafy greens','tomato','cucumber','pumpkin','zucchini','squash','bell pepper','broccoli','cali flower','artichoke','scallion']
flower=['rose','green tea','charmomil','black tea','cactus']
nuts=['almond','peanut','cashew','macadamia','coconut','chia seed','hazelnut','pecan']
fruits=['apple','banana','melon','peach','kiwi','lemon','apricots','plum','straberry','lychees','avocado','pineapple','durian','mango','blueberry','grape','mangosteen']

organ=['legs','arms','wings','necks','thigh','gizzards','intestines','vine','blood vessel','tongue','liver','breast','brain','seeds','knee','bones','hearts','lungs','kidneys','tails','butts','fin','hooves','feet','nails']
other=['yeast','honey','cream','colostrum','lactose','mineral blend','vitamin blend']

seasoning=['salt','surgar','baking soda','baking powder','vinegar','chili','pepper','erythritol','mint','rosemary']

adj=['enriched','dried','dehydrated','bleached','unbleached','fermented','buttered','distilled','cultured']

```

```python
import tracery
from tracery.modifiers import base_english

rules={
    "origin":
    """Ingredients List: #ingredient.capitalize#, #ingredient#, #ingredient#, #ingredient#, #ingredient#, #ingredient#, #ingredient#, #other#, #other#, #other#.""",
    "ingredient":["#meat#","#veg#","#meat# #base#","#veg# #base#","#adj# #meat# #base#","#adj# #veg# #base#","#meat# #organ#","#veg# #organ#"],
    "meat":["#reds#","#flys#","#swims#"],
    "veg":["#starch#","#roots#","#greens#","#flower#","#nuts#","#fruits#"],
    "other":["#adj# #sea#","#adj# #sea# #base#","#sea#","#sea# #base#","#sea# #organ#","#oth#","#meat# #oth#"],
    "base":base,
    "adj":adj,
    "reds":reds,
    "flys":flys,
    "roots":roots,
    "swims":swims,
    "starch":starch,
    "greens":greens,
    "flower":flower,
    "nuts":nuts,
    "fruits":fruits,
    "organ":organ,
    "sea":seasoning, 
    "oth":other
}
grammar = tracery.Grammar(rules)
grammar.add_modifiers(base_english)
grammar.flatten("#origin#")
```

```python
%%capture cap --no-stderr
import textwrap
for i in range(30):
    servingSize=random.randrange(0,100)
    servings=round(random.random()*10,1)
    nutrition=[]
    fat=random.randrange(0,servingSize)
    carbohydrate=random.randrange(servingSize - fat)
    protein=servingSize-fat-carbohydrate
    cholesterol=random.randrange(0,300)
    sodium=random.randrange(0,2300)


    fatPercent=round(fat/44,2)*100
    carbohydratePercent=round(carbohydrate/45,2)*100
    proteinPercent=round(protein/56,2)*100
    sodiumPercent=round(sodium/2300,2)*100
    cholesterolPercent=round(cholesterol/300,2)*100
    calories= 9*fat+4*(carbohydrate+protein)


    # output begins with salutation
    print(color.BOLD +color.UNDERLINE+"Nutrition Lists" + ":"+ color.END)
    print("about " + str(servings) +" servings per container" )
    print("Serving size " + str(servingSize) +"g")
    print(color.BOLD+"Amount per serving")
    print(color.BOLD+color.UNDERLINE +"Calories                  "+ str(calories)+ color.END)
    print("                 "+ "%Daily Value")
    print(color.BOLD+color.UNDERLINE +"Total Fat "+ color.END+color.UNDERLINE + str(fat)+"g"+ color.END)
    print(color.BOLD +"........................."+ str("%.1f" % fatPercent)+"%"+ color.END)
    if fat==0:
        print("  "+"Saturated Fat "+"0"+"g")
        print("  "+"Trans Fat "+"0"+"g")
    else:
        satFat=random.randrange(0,fat)
        print("  "+"Saturated Fat "+str(satFat)+"g")
        print("  "+"Trans Fat "+str(fat-satFat-random.randrange(0,fat-satFat))+"g")

    print(color.BOLD+color.UNDERLINE +"Cholesterol "+ color.END+color.UNDERLINE + str(cholesterol)+"mg"+ color.END)
    print(color.BOLD +"........................."+ str("%.1f" % cholesterolPercent)+"%"+ color.END)

    print(color.BOLD+color.UNDERLINE +"Sodium "+ color.END+color.UNDERLINE + str(sodium)+"mg"+ color.END)
    print(color.BOLD +"........................."+ str("%.1f" % sodiumPercent)+"%"+ color.END)

    print(color.BOLD+color.UNDERLINE +"Total Carbohydrate "+ color.END+color.UNDERLINE + str(carbohydrate) +"g"+ color.END)
    print(color.BOLD +"........................."+ str("%.1f" % carbohydratePercent)+"%"+ color.END)
    if carbohydrate==0:
        print("  "+"Dietary Fiber "+"0"+"g")
        print("  "+"Total Sugars "+"0"+"g")
    else:
        fib=random.randrange(0,carbohydrate)
        print("  "+"Dietary Fiber "+str(fib)+"g")
        ts=carbohydrate-fib-random.randrange(0,carbohydrate-fib)
        print("  "+"Total Sugars "+str(ts)+"g")
        print("    "+"Inlcudes "+str(random.randrange(0,ts))+"g Added Surgars")

    print(color.BOLD+color.UNDERLINE +"Total Protein "+ color.END+color.UNDERLINE+ str(protein)+"g"+ color.END)
    print(color.BOLD +"........................."+ str("%.1f" % proteinPercent)+"%"+ color.END)

    print()
    IL=grammar.flatten("#origin#")
    IL=IL.split()
    
    print(color.BOLD+color.UNDERLINE +IL[0]+IL[1]+ color.END)
    print(textwrap.fill(" ".join(IL[2:]) , 30))
    print()
    print()


    with open('output.txt', 'w') as f:
        f.write(cap.stdout)


```
